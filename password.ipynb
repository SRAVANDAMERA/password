**project password**

```python
import re

passwd = str(input("Enter a password : "))

if len(passwd) != 10 :
  print("Enter a 10 digit alphanumeric password with symbols")

elif re.search(r'[^a-zA-Z0-9]',passwd) and not (passwd.isalnum() or passwd.isdigit()):
  print("Your password is : ",passwd)

else:
  print("Enter a valid password")
```
